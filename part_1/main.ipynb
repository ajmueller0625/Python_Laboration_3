### 21 Använd .apply() 
1. Undersök vilka kategorier det finns i kolumnen "state"
2. Skapa en funktion som baserat på värdena i kolumnen state är, modifierar kolumnen "usd_funding_goal" med en * om det står success, "!" för failure eller canceled, "---" för "suspended".
3. Displaya ett sample av 20 rader från din modifierade kolumn  

```python
# Används för att undersöka vilka värde kolumnen state har
df['state'].unique()

# Funtionen som används för att byta värdena i kolumnen usd_funding_goal
def change_usd_funding_goal_value(old : str) -> str:
    if old == 'successful':
        return '*'
    elif old == 'failed' or old == 'canceled':
        return '!'
    elif old == 'suspended':
        return '---'
    else:
        return old

# Använder funktionen övan för att byta värdena i kolumnen usd_funding_goal som baseras från värdena i kolumnen state
df['usd_funding_goal'] = df['state'].apply(lambda x : change_usd_funding_goal_value(x))

# Hämtar slumpmässigt 20 rader från dataframet
df.sample(20)
```
